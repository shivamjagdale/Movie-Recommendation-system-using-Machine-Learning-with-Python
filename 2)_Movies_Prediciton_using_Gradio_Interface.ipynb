```python
import gradio as gr
import pandas as pd
import difflib
import joblib

# Load your data and similarity matrix
df = pd.read_csv("/content/drive/MyDrive/ML Datasets/movies.csv")  # Make sure this has columns 'title' and 'index'
score = joblib.load("score.pkl")  # or use pickle if you saved it that way

def recommend_movies(name):
    movie_titles = df["title"].tolist()
    close_match = difflib.get_close_matches(name, movie_titles)

    if not close_match:
        return "No close match found. Please try a different movie title."

    final_close_match = close_match[0]
    index_of_close_match = df[df.title == final_close_match]["index"].values[0]

    similarity_score = list(enumerate(score[index_of_close_match]))
    descending = sorted(similarity_score, key=lambda x: x[1], reverse=True)

    recommended_movies = []
    for i, movie in enumerate(descending[1:6], start=1):  # Skip the first because it's the movie itself
        indexx = movie[0]
        moviess = df[df.index == indexx]["title"].values[0]
        recommended_movies.append(f"{i}. {moviess}")

    return "\n".join(recommended_movies)

# Gradio Interface
gr.Interface(
    fn=recommend_movies,
    inputs=gr.Textbox(label="Enter a movie title"),
    outputs=gr.Textbox(label="Recommended Movies"),
    title="🎬 Movie Recommendation System",
    description="Enter a movie you like, and we'll suggest similar ones!"
).launch()
```
